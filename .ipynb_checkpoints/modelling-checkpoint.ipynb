```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
##using a linear regressor to check the score.
tscv = TimeSeriesSplit(n_splits=5,max_train_size=None)
scores = []
for train_index, test_index in tscv.split(X_sub):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y[train_index], y[test_index]
    model = LinearRegression()
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    scores.append(score)
print('mean score: ',np.mean(scores))    
```

#### GradientBoost vs. AdaBoosting

Let's first try to understand what the term 'boosting' actually means and what it does since it literally appears in both types of machine learning approaches. Boosting means using a large number of weak learners that are turned into strong learners and then combined to form the model. In boosting each new tree is fitted on a modified version of the original dataset. The modification process involves using a weak learner to make bad predictions and then weigh poor results such that they have more significance on the next model fitting process.
In adaboosting, an initial weak learner is fitted on a dataset with a weight that is equally distributed to all data points. After fitting this weak learner we increase the weights of those observations that are hard to classify and lower the weights of those that are easier to classify. The second tree is used to fit on the weighted dataset. The idea is to improve the predictions of the first tree. This process is done consecutively until a set number of trees has been used. A final prediction is then the base sum of predictions times applied weight (a weighted sum essentially.).

The splitting point between the two approaches is how the algorithm idetifies the shortcomings of each individual model, while Ada boost rectifies poorly predicted results by an appropriate weight Gradientboost uses a *loss function* The loss function is a measure of how well the coefficients of the learner is able to predict results. One of the biggest motivations to use the gradient boost algorithm is that it gives us a greater sense of control over the process by the option to define the loss function.
